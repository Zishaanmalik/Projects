# ðŸ§  RandomForest Classifier on Digits Dataset

In this notebook, we:
- Load and visualize the **Digits dataset**
- Train Data using **RandomForest Classifier** 
- Check model score

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
# RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
## Step Load Data
data=load_digits()
```

```python
dir(data)
```

```python
x=data.data
```

```python
y=data.target
```

```python
#ploting digits
plt.gray()
for i in range(2):
    plt.matshow(data.images[i])
```

```python
# train and test data
x,tx,y,ty=train_test_split(x,y,test_size=0.2)
```

```python
# creat model and train
model=RandomForestClassifier()
model.fit(x,y)
```

```python
# score of model
model.score(tx,ty)
```

```python
# predict the results
py=model.predict(tx)
```

```python
cm=confusion_matrix(ty,py)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel("truth")
plt.ylabel("pridected")
```
